```python
from __future__ import print_function
import numpy as np
import seaborn as sb
import pylab as pl
%pylab inline
```

```python
#set variables/create dictionaries for each distribution
normDist = ['normal']
normDict = {}
normDict['normal'] = {}

biDist = ['binomial']
biDict = {}
biDict['binomial'] = {}

poisDist = ['poisson']
poisDict = {}
poisDict['poisson'] = {}

chiDist = ['chisq']
chiDict = {}
chiDict['chisq'] = {}

lnormDist = ['lnorm']
lnormDict = {}
lnormDict['lnorm'] = {} 

mymean = 100
# Tested a few different generations of 100 random sizes
#mysize = (2000 / (np.array(range(1, 100)))).astype(int)
mysize = 2000 / (np.array(range(1, 100) + [10]))
np.random.seed(876234)
#mysize3 = (np.random.rand(100) * 2000).astype(int)
print (mysize, mysize.shape)
    #print (mysize2, mysize2.shape)
#print (mysize, mysize.shape)
```

# Normal Distribution

```python
for n in mysize:
    normDict['normal'][n] = np.random.normal(mymean, size = n)

normDict['normal']['means'] = {}
axnorm = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in normDict['normal']:
    if not type(nn) == str:
        normDict['normal']['means'][nn] = normDict['normal'][nn].mean()
        #and plot it
        axnorm.plot(nn, normDict['normal']['means'][nn], 'o')
        axnorm.set_xlabel('Sample Size', fontsize=18)
        axnorm.set_ylabel('Sample Mean', fontsize=18)
        axnorm.set_title('Normal Distribution', fontsize=18)
        axnorm.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### As predicted by the Central Limit Theorem, the means of the 100 samples of normal distributions likewise take on a normal distribution. The range of means is from 0.3 less than to just 0.1 more than the specified mean of 100. Only 3 points are more than 0.1 under 100. Otherwise, the distribution is fairly symmetrical.

# Binomial Distribution

```python
for n in mysize:
    biDict['binomial'][n] = np.random.binomial(200, .5, size = n)
    
biDict['binomial']['means'] = {}
axbinomial = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in biDict['binomial']:
    if not type(nn) == str:
        biDict['binomial']['means'][nn] = biDict['binomial'][nn].mean()
        #and plot it
        axbinomial.plot(nn, biDict['binomial']['means'][nn], 'o')
        axbinomial.set_xlabel('Sample Size', fontsize=10)
        axbinomial.set_ylabel('Sample Mean', fontsize=10)
        axbinomial.set_title('Binomial Distribution', fontsize=10)
        axbinomial.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### The distribution of the means of the samples of the binomial distribution is a little more spreadout than that of the normal distribution, but still approximates a normal distribution. Again, most means are clustered around 100, and the distrinution is fairly symmetrical.

# Chi-Squared Distribution

```python
for n in mysize:
    chiDict['chisq'][n] = np.random.chisquare(mymean, size = n)

chiDict['chisq']['means'] = {}
axchisq = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in chiDict['chisq']:
    if not type(nn) == str:
        chiDict['chisq']['means'][nn] = chiDict['chisq'][nn].mean()
        #and plot it
        axchisq.plot(nn, chiDict['chisq']['means'][nn], 'o')
        axchisq.set_xlabel('Sample Size', fontsize=18)
        axchisq.set_ylabel('Sample Mean', fontsize=18)
        axchisq.set_title('Chi-Squared Distribution', fontsize=18)
        axchisq.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### While a chi-squared distribution itself is not symmetrical, the Central Limit Theorem again holds true when you look at a distribution of 100 samples of the chi-squared distribution.
